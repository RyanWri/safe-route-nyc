# Date and Time Analysis

```python
# Load the dataset
df_date = pd.read_csv(filename, parse_dates=["CRASH DATE"])
```

```python
# Ensure CRASH DATE is in string format for concatenation
df_date["CRASH_DATE_STR"] = df_date["CRASH DATE"].dt.strftime("%Y-%m-%d")

# Combine date and time into one string, then parse to datetime
df_date["CRASH_DATETIME"] = pd.to_datetime(
    df_date["CRASH_DATE_STR"] + " " + df_date["CRASH TIME"]
)
```

```python
# Extract datetime features
df_date["HOUR"] = df_date["CRASH_DATETIME"].dt.hour
df_date["DAY_OF_WEEK"] = df_date["CRASH_DATETIME"].dt.dayofweek  # 0 = Monday
df_date["DAY_NAME"] = df_date["CRASH_DATETIME"].dt.day_name()
df_date["MONTH"] = df_date["CRASH_DATETIME"].dt.month
df_date["YEAR"] = df_date["CRASH_DATETIME"].dt.year
df_date["DATE"] = df_date["CRASH_DATETIME"].dt.date

# Preview results
df_date[
    ["CRASH_DATE_STR", "CRASH TIME", "CRASH_DATETIME", "HOUR", "DAY_NAME", "MONTH"]
].head()
```

```python
# Plot: Crashes by Hour of Day
plt.figure(figsize=(10, 6))
df_date["HOUR"].value_counts().sort_index().plot(kind="bar")
plt.title("Crash Frequency by Hour of Day")
plt.xlabel("Hour (0â€“23)")
plt.ylabel("Number of Crashes")
plt.xticks(rotation=0)
plt.grid(axis="y")
plt.tight_layout()
plt.show()
```

```python
# Plot: Crashes by Day of Week
plt.figure(figsize=(10, 6))
day_order = [
    "Monday",
    "Tuesday",
    "Wednesday",
    "Thursday",
    "Friday",
    "Saturday",
    "Sunday",
]
df_date["DAY_NAME"].value_counts().reindex(day_order).plot(kind="bar")
plt.title("Crash Frequency by Day of the Week")
plt.xlabel("Day")
plt.ylabel("Number of Crashes")
plt.xticks(rotation=45)
plt.grid(axis="y")
plt.tight_layout()
plt.show()
```

```python
# Plot: Crashes by Month
plt.figure(figsize=(10, 6))
month_names = {
    1: "Jan",
    2: "Feb",
    3: "Mar",
    4: "Apr",
    5: "May",
    6: "Jun",
    7: "Jul",
    8: "Aug",
    9: "Sep",
    10: "Oct",
    11: "Nov",
    12: "Dec",
}
df_date["MONTH"].map(month_names).value_counts().reindex(month_names.values()).plot(
    kind="bar"
)
plt.title("Crash Frequency by Month")
plt.xlabel("Month")
plt.ylabel("Number of Crashes")
plt.xticks(rotation=0)
plt.grid(axis="y")
plt.tight_layout()
plt.show()
```

# Location Analysis

```python
# Create location-based DataFrame
location_features = [
    "LATITUDE",
    "LONGITUDE",
    "ZIP CODE",
    "ON STREET NAME",
    "CROSS STREET NAME",
    "OFF STREET NAME",
]
df_location = df[location_features].copy()

# Preview relevant columns
df_location.head()
```

```python
# Combine ON, CROSS, and OFF STREET names for total crashes per street
street_counts = pd.concat(
    [
        df_location["ON STREET NAME"],
        df_location["CROSS STREET NAME"],
        df_location["OFF STREET NAME"],
    ]
)

# Count crashes per street and filter
top_streets = street_counts.value_counts().dropna().head(15)

# Plot: Top 15 crash-prone streets
plt.figure(figsize=(10, 6))
top_streets.plot(kind="barh")
plt.title("Top 15 Streets with Most Crashes")
plt.xlabel("Number of Crashes")
plt.ylabel("Street Name")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
# Drop null ZIPs and convert to int
df_location = df_location.dropna(subset=["ZIP CODE"])
df_location["ZIP CODE"] = df_location["ZIP CODE"].astype(int)

# Plot: Crash count by ZIP code
plt.figure(figsize=(12, 6))
df_location["ZIP CODE"].value_counts().sort_values(ascending=False).head(15).plot(
    kind="bar"
)
plt.title("Crashes by ZIP Code (Top 15 in Manhattan)")
plt.xlabel("ZIP Code")
plt.ylabel("Number of Crashes")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Plot: Hexbin for all crash points
plt.figure(figsize=(10, 8))
plt.hexbin(
    df_location["LONGITUDE"],
    df_location["LATITUDE"],
    gridsize=100,
    cmap="Reds",
    mincnt=1,
)
plt.title("Crash Density by Location (Hexbin) â€“ Manhattan")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.colorbar(label="Crash Count")
plt.tight_layout()
plt.show()
```
